# PatchTSMixer Visualization

Visualize results from the PatchTSMixer backtest.

Run this after `patchts_colab.ipynb`.

```python
import os

import pandas as pd

from src.visualization import plot_diagnostic_scatter, plot_residual_histogram, plot_timeseries_forecast

MODEL_NAME = "PatchTSMixer"
COLOR = "#1f77b4"
RESULTS_DIR = "results_patchts"
csv_path = os.path.join(RESULTS_DIR, "patchts_results.csv")

results = pd.read_csv(csv_path)
print(f"Loaded {len(results)} predictions from {csv_path}")
results.describe()
```

```python
plot_timeseries_forecast(results, MODEL_NAME, COLOR)
```

```python
plot_diagnostic_scatter(results, MODEL_NAME, COLOR)
```

```python
plot_residual_histogram(results, MODEL_NAME, COLOR)
```

## Training Loss Curves

Visualize per-epoch training losses across chunks.
Requires running the backtest with `--loss-log-path`.

```python
from glob import glob

from src.visualization import plot_training_losses

loss_files = sorted(glob(os.path.join(RESULTS_DIR, "training_losses*.csv")))
if loss_files:
    for lf in loss_files:
        print(f"Plotting losses from {lf}")
        plot_training_losses(lf, model_name=MODEL_NAME)
else:
    print("No training loss files found. Run backtest with --loss-log-path to generate them.")
```
